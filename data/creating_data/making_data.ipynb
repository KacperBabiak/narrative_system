```python
def flatten_lists(l):
    final = []
    
    for x in l:
        
        smaller_list = []
        for y in x:
            
            if type(y) is list:
                smaller_list.extend(y)
            else: smaller_list.append(y)
            
            final.append(smaller_list)
    return final
    
```

```python
def get_results(row):
    plan = row.results
    character = row.active
    
    actions = plan.split(') ')
    for action in actions:
        
        if 'get_item' in action:
            #print("item!")
            items = action.split('get_item({}, '.format(character))[1].split(' ')[1].replace(')', '').rstrip()
            row['wants_'+items] = 1

            #print( items)
        elif 'attack(doctor' in action:
            #print("attack!")
            chars = action.split('attack({}, '.format(character))[1].split(' ')[0].replace(')', '').rstrip()
            row['attacks_'+chars] = 1
            #print( chars)
    
    return row
        
```

```python
dfc = df2.copy()
```

```python
new_data = {'wants_meds' : None, 'wants_food' : None, 'attacks_doctor' : None, 'attacks_soldier' : None, 'attacks_actor' : None}
dfc = dfc.assign(**new_data)
```

```python
dfc = dfc.apply(get_results,axis=1)
```
