```python
def check_range(val, min, max):
    
    sign, num = val.split(' ')
    l  = [] 

    if sign == '>':
        l= ([i for i in range(int(num)+1, max+1)]  )
    elif sign == '>=':
        l = ([i for i in range(int(num), max+1)])    
    elif sign == '<':
        l=  [i for i in range(min,int(num))]
    elif sign == '<=':
        l= [i for i in range(min,int(num)+1)]
    
    return l
        
   
```

```python
for index, row in df[df.module == main_module].iterrows():
    chosen_actions.append(row['action_name'])
    
    altruism_val = check_range(row['altruism'],-2,2)
    ambition_val = check_range(row['ambition'],0,4)
    
    comb = itertools.product(altruism_val,ambition_val)
    for i in comb:
        if i in dic_key.keys():
            
            dic_key[i] = 1
        
```

```python
print([ i for i in dic_key.keys()])
```

```python
while 0 in dic_key.values():
    random_idx = random.randint(0,len(df))
    row = df.iloc[random_idx]

    altruism_val = check_range(row['altruism'],-2,2)
    ambition_val = check_range(row['ambition'],0,4)
    
    comb = itertools.product(altruism_val,ambition_val)
    for i in comb:
        
        if i in dic_key.keys():
            
            if dic_key[i] == 0:
                chosen_actions.append(row['action_name'])
                dic_key[i] = 1

chosen_actions= list(set(chosen_actions))
```

```python
df_req = pd.read_csv('requirements.csv',index_col='index')
df_req

requirements = {}
columns = df_req.columns
for index,row in df_req.iterrows():
    for column in columns:
        name = str(index) + "_" + column
        
        requirements[name] = df_req.loc[index,column]

print(requirements)
```

```python
def key_action_to_text(row):
    entities = ""
    for ent in row.entities.split(';'):
        
        if 'char' in ent:
            entities = entities + ent + ":character,"
        elif 'item' in ent:
            entities = entities + ent + ":item,"
    entities = entities.rstrip(',')

    req = ""
    items = ['food', 'weapon', 'book', 'poison', 'money' ,'item']
    for r in row.requires.split(';'):
        
        req = req +  str(requirements[r.replace(' ','')]) + " & "
        
    req = req.rstrip(' & ')

    altruism = row.altruism.split(';')
    ambition = row.ambition.split(';')

    sat = "if ("
    sat = sat + 'altruism(char) ' + altruism[0]
    if len(altruism) > 0:
        sat = sat + '& altruism(char) ' + altruism[1]
    
    sat = sat + 'ambition(char) ' + ambition[0]
    if len(ambition) > 0:
        sat = sat + 'ambition(char) ' + ambition[1]
    
    sat += ") \n            satisfaction(char) = satisfaction(char) + 1 "

    text= f"""
    action {row.action_name}({entities}) {{
        precondition:
    
            {req}
            ;
        effect:
        {sat}
            ;
        consenting: {row.consents};
    }};

    """
    return text
```
